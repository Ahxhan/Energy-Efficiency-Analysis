```r
# Creating a new column 'Scaled_Surface_Area'
energy_data <- mutate(data_csv, Scaled_Surface_Area = scale(Surface Area))
print(head(energy_data))
```

```r
# Summarizing average Heating Load and Cooling Load by Orientation
summary_data <- data_csv %>%
  group_by(`Orientation`) %>%
  summarize(
    Avg_Heating_Load = mean(`Heating Load`),
    Avg_Cooling_Load = mean(`Cooling Load`)
  )
print(summary_data)
```

```r
# Load ggplot2 package
library(ggplot2)

# Create the plot
ggplot(energy_data, aes(x = `Relative Compactness`, y = `Heating Load`)) +
  geom_point() +
  ggtitle("Relative Compactness vs. Heating Load") +
  xlab("Relative Compactness") +
  ylab("Heating Load")
```

```r
ggplot(energy_data, aes(x = `Relative Compactness`, y = `Heating Load`)) +
  geom_point() +
  ggtitle("Relative Compactness vs. Heating Load") +
  xlab("Relative Compactness") +
  ylab("Heating Load")
```

```r
ggplot(data_csv, aes(x = `Surface Area`)) +
  geom_histogram(binwidth = 50, fill = "blue", color = "black") +
  ggtitle("Distribution of Surface Area") +
  xlab("Surface Area") +
  ylab("Frequency")
```

```r
ggplot(energy_data, aes(x = factor(`Orientation`), y = `Cooling Load`)) +
  geom_boxplot() +
  ggtitle("Cooling Load by Orientation") +
  xlab("Orientation") +
  ylab("Cooling Load")
```
